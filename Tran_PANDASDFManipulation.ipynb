# PANDAS DataFrame Manipulation

```python
import pandas as pd
nba = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv") 
nba.head()
```

 Adding a new colum named "College Degree" right before the column "College". Filling the rows of this column with "True" if there is a college name entry under the column "College" for the associated row. 

```python
nba.loc[nba['College'].isnull(), 'College'] = 0 
nba_new = nba
nba_new['College Degree'] = [False if x == 0 else True for x in nba_new['College']]
nba_new.head()
```

```python
#confirming this column is boolean
nba_new['College Degree'].dtype
```

Creating a list that holds the names of the players who do not have any college degree (or no university name entry given under "College" column in the data set) through using a for loop and if statement. 

```python
#players who do not have any college degree
nd = []
i = 0
for i, row in nba_new.loc[nba_new['College Degree'] == False].iterrows():
    myTuple = (i,row)
    nd.append(myTuple)
    i += 1
nba_nd = pd.DataFrame(nd)
nba_nd.columns = ['Number','Name']
nba_nd.head()
```

Creating two dataframes named "no_college_degree" and "with_college_degree" with names and salaries of the players in the corresponding category (no college/college)

```python
#no college degree and salary
ncd = []
ncds = []
no_college_degree = pd.DataFrame()

for i, row in nba_new[nba_new['College'] == 0].iterrows():
    ncd.append(row['Name'])
    ncds.append(row['Salary'])

no_college_degree['Name'] = ncd
no_college_degree['Salary'] = ncds
no_college_degree.head()
```

```python
#with college degree
wcd = []
wcds = []
with_college_degree = pd.DataFrame()

for i, row in nba_new[nba_new['College']!= 0 ].iterrows():
    wcd.append(row['Name'])
    wcds.append(row['Salary'])

with_college_degree['Name'] = wcd
with_college_degree['Salary'] = wcds
with_college_degree.head()
```

Reporting the average salary of platers with college and no college degree

```python
no_college_degree['Salary'].mean()
```

```python
with_college_degree['Salary'].mean()
```

Renaming column named "Height" to "Height (feet-inches)". Adding a new column named "Height (inches)" to the existing dataframe. The column should include values in total inches. <br/>
Adding another column named "Tall" to the existing dataframe. This column indicates a bool type value of True if the associated player height is greater than or equal to 80 inches. Otherwise, its value is False. 

```python
nba_new.rename(columns={'Height': 'Height (feet-inches)'}, inplace = True)
nba_new.head()
```

```python
#splitting the height column by the '-' to separate feet from inches.
new_nba_new = nba_new["Height (feet-inches)"].str.split("-", n = 1, expand = True)
new_nba_new.head()
```

```python
#replacing the Height (feet-inches) column with column 0 from previously split values.
nba_new["Height (feet-inches)"]= new_nba_new[0] 
#creating a new inches column that represents the 1 column from previously split DF
nba_new["Inches"]= new_nba_new[1] 
```

```python
#dropping NaNs and converting new columns into integers
nba_new.dropna(inplace = True) 
nba_new["Height (feet-inches)"]= nba_new["Height (feet-inches)"].astype(str).astype(int)
nba_new["Inches"]= nba_new["Inches"].astype(str).astype(int)
```

```python
#multiplying feet by 12 and adding inches column to get height in inches
nba_new["Height (Inches)"] = (12 * nba_new["Height (feet-inches)"]) + (nba_new["Inches"])
nba_new.head()
```

```python
#creating a new column that represents a boolean value of true if the player is taller than 80 inches.
nba_new['Tall'] = [True if x >= 80 else False for x in nba_new['Height (Inches)']]
nba_new.head()
```

By using the dataframe with the new column "Tall", report the agerage salaries of players who are labeled as 'True' under Tall column. Also, report the average salaries of players who are labeled as 'False' under Tall column.<br/>

```python
#Salaries for those labeled 'True' (gt 80 inches)
#sts = super tall salaries
sts = []

super_tall_salaries = pd.DataFrame()
for i, row in nba_new[nba_new['Tall'] == True].iterrows():
        sts.append(row['Salary'])

super_tall_salaries['Salary'] = sts
super_tall_salaries.head()
```

```python
super_tall_salaries['Salary'].mean()
```

```python
#reporting salaries for those labelled false (lt 80 inches)
#lts = less tall salaries
lts = []

less_tall_salaries = pd.DataFrame()
for i, row in nba_new[nba_new['Tall'] == False].iterrows():
        lts.append(row['Salary'])

less_tall_salaries['Salary'] = lts
less_tall_salaries.head()
```

```python
less_tall_salaries['Salary'].mean()
```

Adding a new column named "Young" to the existing dataframe. This column indicates a bool type value of True if the associated player age is less than 30. Otherwise, its value is be False. 

```python
nba_new['Young'] = [True if x < 30 else False for x in nba_new['Age']]
nba_new.head()
```

Using the dataframe with the new column "Young", report the average salaries of players who are labeled as 'True' under Young column. Also, report the average salaries of players who are labeled as 'False' under Young column.

```python
#salaris for those labeled as 'True'
#ys = young salaries
ys = []

young_salaries = pd.DataFrame()
for i, row in nba_new[nba_new['Young'] == True].iterrows():
        ys.append(row['Salary'])

young_salaries['Salary'] = ys
young_salaries.head()
```

```python
young_salaries['Salary'].mean()
```

```python
#reporting for those labeled as 'false'
#lys =  less young salaries
lys = []

less_young_salaries = pd.DataFrame()
for i, row in nba_new[nba_new['Young'] == False].iterrows():
        lys.append(row['Salary'])

less_young_salaries['Salary'] = lys
less_young_salaries.head()
```

```python
less_young_salaries['Salary'].mean()
```
